# merge desa and kecamatan

```python
import pandas as pd
import numpy as np

data = pd.read_csv('m_desa_jatim.csv')
data['KECAMATAN'] = data['KECAMATAN'].astype(str)

data_kecamatan = pd.read_csv('m_kecamatan.csv')
data_kecamatan['KODE'] = data_kecamatan['KODE'].astype(str)

data_gabungan_kecamatan_desa = pd.merge(data, data_kecamatan, left_on='KECAMATAN', right_on='KODE', how='left', suffixes=('_desa','_kecamatan'))
```
